```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 1, 1000)
y = np.linspace(0, 1, 1000)
X, Y = np.meshgrid(x, y)

Re = 40
viscosity = 1 * max(y) / Re

lmbda = 0.5 *(Re-np.sqrt(Re**2 + 16 * np.pi**2))
print (lmbda)
# lmbda = 1/2*viscosity - np.sqrt (1/4*viscosity**2 + 4*np.pi**2)

Vx = 1 - np.exp(lmbda * X) * np.cos(2 * np.pi * Y)
Vy = (lmbda / (2 * np.pi)) * np.exp(lmbda * X) * np.sin(2 * np.pi * Y)

P = 0.5 * (1-np.exp(2*lmbda*X))
```

```python
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Vx, levels=50, cmap='plasma')  

cbar = plt.colorbar(contour)
cbar.set_label('Значения Vx')  

plt.title('График Vx')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, Vy, levels=20, cmap='plasma') 

cbar = plt.colorbar(contour)
cbar.set_label('Значения Vy')  

plt.title('График Vy')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.figure(figsize=(8, 6))
contour = plt.contourf(X, Y, P, levels=20, cmap='viridis') 

cbar = plt.colorbar(contour)
cbar.set_label('Значения P') 

plt.title('Распределение давления')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```

```python
plt.streamplot(X, Y, Vx, Vy, density=2, color='b')
plt.title('Линии тока (поток Ковашнаи)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
step = 35
plt.quiver(X[::step, ::step], Y[::step, ::step], 
           Vx[::step, ::step], Vy[::step, ::step], 
           scale=5, 
           color='blue',  
           width=0.003,   
           headwidth=3,   
           headlength=5)  
plt.title('Векторное поле скорости (поток Ковашнаи)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
